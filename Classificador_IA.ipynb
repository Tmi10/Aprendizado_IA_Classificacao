```python
# Importar Bilbiotecas de Manipulação de Dados 

import numpy as np
import pandas as pd
```

```python
# Comando para ler arquivo csv 

df = pd.read_csv('data_cancer2.csv', encoding='utf-8')
```

```python
# Ver os primeiros dados do dataframe

df.head()
```

```python
# Tirar uma cópia do dataframe

df2 = pd.DataFrame.copy(df)
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
# Use o método drop para deletar a coluna

df2 = df2.drop('id', axis=1)
```

```python
df2.head()
```

```python
# Use o método dropna para eliminar colunas onde todos os valores são None

df2 = df2.dropna(axis=1, how='all')
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
# Ver a quantidade de linha e colunas 

df2.shape
```

```python
# Verificar se existe alguma coluna nula

df2.isnull().sum()
```

SEPARAR OS DADOS EM PREVISORES E ALVO, ASSIM COMO TRANSFORMAR OS ATRIBUTOS CATEGÓRICOS EM NUMÉRICOS

```python
previsores = df2.iloc[:, 1:32].values

previsores
```

```python
previsores.shape
```

```python
alvo = df2.iloc[:, 0].values

alvo
```

```python
# Transformar os dados categóricos do Alvo em numérico

from sklearn.preprocessing import LabelEncoder

alvo = LabelEncoder().fit_transform(alvo)
```

```python
alvo
```

```python
alvo.shape
```

```python
# Escalonamento dos previsores

from sklearn.preprocessing import StandardScaler

previsores_esc = StandardScaler().fit_transform(previsores)
```

```python
previsores_esc
```

```python
df2.describe()
```

SEPARAR A BASE DE DADOS EM TREINO E TESTE

```python
#Biblioteca de Teste e Treino

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, alvo, test_size = 0.3, random_state = 0)
```

```python
x_treino.shape
```

```python
x_teste.shape
```

```python
y_treino.shape
```

```python
y_teste.shape
```

NAIVE BAYES

```python
# Biblioteca Naive Bayes

from sklearn.naive_bayes import GaussianNB

naive = GaussianNB()

naive.fit(x_treino, y_treino)
```

```python
# Avaliação do Algoritmo

previsoes_naive = naive.predict(x_teste)

previsoes_naive
```

```python
y_teste
```

```python
# Biblioteca para descobrir a acurácia, matriz de confusão e classificação do algoritmo

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Acurácia

accuracy_score(y_teste, previsoes_naive)
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_naive) * 100.0))
```

```python
# Matriz de Confusão

confusion_matrix(y_teste, previsoes_naive)
```

```python
# Avaliação do Modelo Teste

print(classification_report(y_teste, previsoes_naive))
```

```python
# Análise dados de treino

previsoes_treino_naive = naive.predict(x_treino)

previsoes_treino_naive
```

```python
accuracy_score(y_treino, previsoes_treino_naive)
```

```python
confusion_matrix(y_treino, previsoes_treino_naive)
```

```python
# Bibliotecas Validação Cruzada 

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# Separando os dados em folds

kfold = KFold(n_splits = 30, shuffle=True, random_state = 5)
```

```python
# Criando o modelo

modelo_nb = GaussianNB()

resultado_nb = cross_val_score(modelo_nb, previsores, alvo, cv = kfold)

resultado_nb
```

```python
# Usamos a média e o desvio padrão

print("Acurácia Média: %.2f%%" % (resultado_nb.mean() * 100.0))
```

LIGHTGBM

```python
import lightgbm as lgb

# Dataset para treino
dataset = lgb.Dataset(x_treino,label=y_treino)

# Parâmetros
parametros = {'num_leaves':250, # número de folhas
              'objective':'binary', # classificação Binária
              'max_depth':2,
              'learning_rate':.05,
              'max_bin':100}

```

```python
# Marcação do tempo de execução
from datetime import datetime
inicio=datetime.now()
lgbm=lgb.train(parametros,dataset)
fim=datetime.now()

tempo = fim - inicio
tempo
```

```python
previsoes_lgbm = lgbm.predict(x_teste)
previsoes_lgbm
```

```python
previsoes_lgbm.shape
```

```python
# Quando for menor que 5 considera 0 e quando for maior ou igual a 5 considera 1
for i in range(0, 171):
    if previsoes_lgbm[i] >= .5:
       previsoes_lgbm[i] = 1
    else:
       previsoes_lgbm[i] = 0
```

```python
previsoes_lgbm
```

```python
y_teste
```

```python
print("Acurácia: %.2f%%" % (accuracy_score(y_teste, previsoes_lgbm) * 100.0))
```

```python
confusion_matrix(y_teste, previsoes_lgbm)
```

```python
# Análise dados de treino

previsoes_treino_lgbm = lgbm.predict(x_treino)
previsoes_treino_lgbm
```

```python
previsoes_treino_lgbm.shape
```

```python
# Quando for menor que 5 considera 0 e quando for maior ou igual a 5 considera 1
for i in range(0, 398):
    if previsoes_treino_lgbm[i] >= .5:
       previsoes_treino_lgbm[i] = 1
    else:
       previsoes_treino_lgbm[i] = 0
```
